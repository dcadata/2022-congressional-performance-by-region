# 2022 Congressional Performance by Region

**Comparing congressional performance by region in the 2022 elections**

```python
import task  # import data processing functions already created
```

```python
results = task.read_and_merge_all()
```

## Democratic performance in NY and CA

Excluding uncontested seats

---

After accounting for incumbency, NY/CA Congressional Democrats underperformed Biden by 8.3 points:

```python
regions = 'NY CA'
df = task.filter_by_region(results.copy(), regions.split())
summary = task.make_plot_and_fit_model(df, 'votePct ~ votePctPres + Location * isIncumbent')
print(summary)
```

In competitive districts, they underperformed Biden by 3.7 points:

```python
summary = task.make_plot_and_fit_model(df, 'votePct ~ votePctPres + Location * isIncumbent', competitive_only=True)
print(summary)
```
